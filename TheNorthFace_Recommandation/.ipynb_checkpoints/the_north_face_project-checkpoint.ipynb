```python
# -*- coding: utf-8 -*-
"""the_north_face_project.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1TgCrVcJjXgN7GMaBdUhQvFzTd8CZyQOG

<h1> THE NORTH FACE PROJECT </h1>
"""



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

import spacy
from wordcloud import WordCloud

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN
from sklearn.metrics.pairwise import cosine_distances, cosine_similarity
from sklearn.decomposition import TruncatedSVD

df= pd.read_csv('C:/Users/serda/OneDrive/Bureau/Online Education/Certification/TheNorthFace_Recommandation/sample-data.csv')

df.head()

df.info()

df.describe(include='all')

df.head()

df_clean= df.copy() ### We store the original dataset intact for further comparasions and uses

nlp = spacy.load("en_core_web_sm")

def clean_text(text):
    text = nlp(text)
    cleaned_text = []
    for token in text:
        if not token.is_stop: # Ignore stopwords no need to import STOP_WORDS because actual code consist of default Stop Words
            lemma = token.lemma_ # token.text is not necessary in our case because we need lemmatized version of words. 
            if lemma == "'s": # handle contractions separately
                lemma = "is"
            elif lemma == "'ll":
                lemma = "will"
            elif lemma == "'ve":
                lemma = "have"
            elif lemma == "'re":
                lemma = "are"
            cleaned_text.append(lemma)
    cleaned_text = " ".join(cleaned_text)
    cleaned_text = cleaned_text.lower()
    cleaned_text = ''.join(c for c in cleaned_text if c.isalnum() or c.isspace())
    return cleaned_text

df_clean['product_name'] = df_clean.description.apply(lambda x: x.split('-')[0])
df_clean['cleaned_descriptions'] = df_clean.description.apply(lambda k: ' '.join(k.split('-')[1:]).strip())
 ### I prefered to create this column with the names because product_name itself can be important in classifications

df_clean.head()

# Cleaning our df_clean copy set with above function: 
df_clean['cleaned_descriptions'] = df_clean['description'].apply(clean_text)

df_clean.head()

print ((df_clean.cleaned_descriptions)[0])

### thought that %100 cotton etc... maybe important . Then decided to keep all numeric values.

###Vectorizing

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(df_clean['cleaned_descriptions'])
feature_names = vectorizer.get_feature_names_out()

#here is a sample of vectorisation . This is a sparse matrix . Mostly consist of zeros . 
#After fitting only the most important words will be ordered. Over 4935 unique words
list(tfidf_matrix)[:3]
# We will not convert it to np.array. Because many of scikit-learn algoritmes are good at sparse matrixes. 
#tfidf_matrix = tfidf_matrix.toarray()

tfidf_matrix.shape

tfidf_matrix

"""***MODEL DBSCAN***




"""

distance_c = 1 - cosine_distances(tfidf_matrix) ### cosine distances is between 0-1-2 while cosine similarity are between 1-0 and -1
#these are the vectors , which are identical, no linked, or opposites 

# DBSCAN
db = DBSCAN(eps=0.2, min_samples=4, metric="cosine") #no need to suggest algoritme because we did not choose euclidien
db.fit(distance_c) ##TF-IDF values are already normalized between 0 and 1 #precomputed metric fits here.
#we should have used precomputed because classifications metrics does not support cosine distances with eps

# Print out the labels - these are the cluster assignments for each row in the matrix
print(db.labels_)

db.labels_

from collections import Counter

# Suppose 'labels' is your list of cluster assignments from DBSCAN
labels = db.labels_

# Use Counter to count occurrences of each label
label_counts = Counter(labels)

# Print the size of each cluster
for cluster_label, count in label_counts.items():
    print(f"Cluster {cluster_label}: {count} products")

documents = df_clean['cleaned_descriptions'].tolist()
labels = db.labels_ 
labels_no_outliers = db.labels_[db.labels_ > -1]


db_clusters = {}

for label, document in zip(labels_no_outliers, documents): 
    if label not in db_clusters:  ##if not in db_clusters then append
        db_clusters[label] = document ### otherwise open a new document with this cluster_text
    else:
        db_clusters[label] += " " + document

n_clusters = max(db.labels_) + 1  # get number of clusters
n_cols = 4  # set number of columns for subplots
n_rows = n_clusters // n_cols + (n_clusters % n_cols > 0)  # calculate number of rows needed 
#(n_clusters % n_cols > 0) checks if there is a remainder

fig, axs = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 5))  # create subplots

for i, (label, text) in enumerate(db_clusters.items()):
    if label != -1:  # exclude outliers
        row = i // n_cols
        col = i % n_cols
        wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(text)
        axs[row, col].imshow(wordcloud, interpolation='bilinear')
        axs[row, col].axis('off')
        axs[row, col].set_title(f"Word Cloud for Cluster {label}")

# Hide empty subplots
if n_clusters % n_cols != 0:
    for col in range(n_clusters % n_cols, n_cols):
        fig.delaxes(axs[n_rows - 1, col])

plt.tight_layout() 
plt.show()



"""**Recommandation System**"""

cosine_sim = cosine_similarity(tfidf_matrix)

def find_similar_items(item_id, df_clean, cosine_sim, max_recommend=5):
    # Get the index of the item that matches the item_id
    idx = df_clean[df_clean['id'] == item_id].index[0]

    # Get the pairwise similarity scores of all items with that item
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the items based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the most similar items
    sim_scores = sim_scores[1:max_recommend+1]  # we start from 1 to exclude the item itself

    # Get the item indices
    item_indices = [i[0] for i in sim_scores]

    # Return the top most similar items
    return df_clean['id'].iloc[item_indices].values

item_id = int(input(" Please enter the product name: "))
recommended_items = find_similar_items(item_id, df_clean, cosine_sim)

print("You may also be interested in these products: ")
for item in recommended_items:
    print(f"Product ID: {item}, Product Name: {df_clean.loc[df_clean['id'] == item, 'product_name'].values[0]}")


df_clean.columns



"""## **TOPIC**"""

n_components = 20  
svd_model = TruncatedSVD(n_components=n_components, algorithm='randomized', n_iter=100, random_state=42)

lsa = svd_model.fit_transform(tfidf_matrix)

topic_encoded_df = pd.DataFrame(lsa, columns=[f'topic_{i+1}' for i in range(n_components)], index = df_clean.index)
topic_encoded_df["text"] = df_clean['cleaned_descriptions'].values

topic_encoded_df["class_pred"] = [np.argmax(topic) for topic in lsa]
topic_encoded_df['main_topic'] = topic_encoded_df.drop('text', axis=1).idxmax(axis=1)

print(topic_encoded_df["class_pred"].value_counts())
print(topic_encoded_df["main_topic"].value_counts())

topic_encoded_df.head()

n_topics = 20  # number of topics
n_cols = 4  # set number of columns for subplots
n_rows = n_topics // n_cols + (n_topics % n_cols > 0)  # calculate number of rows needed 

n_top_words = 10

fig, axs = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 5))  # create subplots
axs = axs.flatten()  # flatten the array

for i in range(n_topics):
    component = svd_model.components_[i]
    top_word_indices = component.argsort()[-n_top_words:]
    top_words = [vectorizer.get_feature_names_out()[idx] for idx in top_word_indices]
    wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(' '.join(top_words))
    axs[i].imshow(wordcloud, interpolation='bilinear')
    axs[i].axis('off')
    axs[i].set_title(f"Word Cloud for Topic {i+1}")

# Hide empty subplots
for i in range(n_topics, n_rows*n_cols):
    fig.delaxes(axs[i])

plt.tight_layout() 
plt.show()

import sklearn
print(sklearn.__version__)

```
